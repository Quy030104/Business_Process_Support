```python
data['City']
data.replace('NYC','New York City', inplace = True)
data
```

```python
data.dropna(inplace=True)
data
```

```python
# Phan tich du lieu
quantities = data["Quantity Ordered"]
quantities.sum()
quantities.mean()

prices = data["Price Each"]
prices.mean()
prices.median()

# Them moi cot du lieu
data["Detail Address"]= data['City']+''+data['Address']
data
data["Sale"] = data["Price Each"] * data["Quantity Ordered"]
data.describe()
```

```python
import matplotlib.pyplot as plt

# ThÃªm cá»™t 'Month' vÃ o DataFrame data
data["Month"] = data["Date"].str[0:2]

# TÃ­nh tá»•ng doanh sá»‘ theo thÃ¡ng
sales = data.groupby("Month").sum()["Sale"]

# Váº½ biá»ƒu Ä‘á»“
plt.figure(figsize=(10, 6))
sales.plot(kind='bar', color='pink')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# TÃ¬m thÃ¡ng cÃ³ doanh sá»‘ cao nháº¥t
best_month = sales.idxmax()
best_month_sales = sales.max()
print(f'The best month for sales is {best_month} with total sales of ${best_month_sales}.')

# Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
plt.show()
```

```python
df =pd.DataFrame(data)
city_sales = data.groupby('City').sum()['Sale']
plt.figure(figsize=(10, 6))
city_sales.plot(kind='bar', color='pink')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
best_city = city_sales.idxmax()
best_city_sales=city_sales.max()
print(f'The best city with the highest sales {best_city} with total sales of ${best_city_sales}.')
plt.show()
```

```python
N = 10
store_sales=df.groupby('StoreID')['Sale'].sum()
top_nstores=store_sales.nlargest(N)
bottom_nstores = store_sales.nsmallest(N)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
top_nstores.plot(kind='bar',color='pink')
plt.title(f'Top {N} StoreID by Sales')
plt.xlabel('StoreID')
plt.ylabel('Total Sales')
```

```python
plt.subplot(1,2,2)
bottom_nstores.plot(kind='bar', color='pink')
plt.title(f'Bottom {N} StoreIDs by Sales')
plt.xlabel('StoresID')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
